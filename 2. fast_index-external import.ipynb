## selecting data in pandas

### Using iloc to select by position

a dataFrame can be represented as a simple table or A 2D Array. you can use iloc to select rows and cols in the table by their position.

iloc syntax is easy:
df.iloc[row_axis,col_axis]
for example:
the following snippet selects the first row and 2nd column

````python
df.iloc[0,1]
````

You can use a slicing notation similar to numpy with iloc
<br>
For example:
<br>
In the next snippet we choose the last ten columns for all rows

````python
df.iloc[:,-10:].head() #head() is used to show only the 5 rows from the selection to make the output fit nicely in the noebook
````

you can learn more about the slicing notation in https://www.tutorialspoint.com/numpy/numpy_indexing_and_slicing.htm

### iloc is easy but...

some queries with iloc could be a real challange, what if you want to select rows by a certain value which exists in the data e.g age or club?
<br>
one could use boolean indexing, which means passing an array of boolean values in the length if the data frame to the iloc selector.
<br>
every value in the array which results to True will be selected.

we can create an array of values by simply comparing a column with some value e.g

````python
(df.Age<32).values
````

now we can pass the array to the iloc selector and get the selected data

````python
df.iloc[(df.Age<32).values].head()
````

but what if the we want another condition e.g age higher then 27?
<br>
we could just do:

````python
df.iloc[(df.Age<32).values&(df.Age>27).values].head()
````

or on more specific cases:

````python
df.iloc[(df.Age<32).values&(df.Age>27).values&(df.Nationality=="Argentina").values&(df.Club=="FC Barcelona").values,1]
````

### loc

selecting using loc is done by using 2 values, the **index** and the **columns**.
<br>
but this time the index is not the same position wise index we used in the iloc.
<br>
this time the index is what you say it is!

the predefined index in pandas when reading csv is a lot like the position index used by iloc. lets examine out DataFrame default index

````python
df.head()
````

see the 0 to one column with no name on it?
this is the index of the DataFrame.
we can select using loc with index and column

````python
df.loc[0:3,"Age"]
````

notice that the notation is 
```python 
df.loc[index,column]
```

you can also pass an array of indexes and columns

````python
df.loc[[0,4,22,50],["Age","ID"]]
````

boolean indexing is also possible for example:

````python
df.loc[(df.Age<32)&(df.Age>23)].head()
````

in the example above we did not choose any column which default to selecting all the columns this is equivalent to:


````python
df.loc[(df.Age<32)&(df.Age>23),:].head()
````

#### hey but what is so spacial about loc if the index and the position are the same?
the answer is that you can choose the index to be any value you like! 

you can set an index by a column like this:

````python
df.set_index(["Club"],inplace=True,drop=False)
````

notice the `drop=False` passed to `df.set_index()` which keeps the column in the DataFrame and in the index <br>

you can now select players by club!

````python
df.loc["Chelsea"].head()
````

some more examples:

````python
df.loc[["Chelsea","FC Barcelona"]]
````

````python
df.loc[["Chelsea","FC Barcelona"],["Age"]]
````

#### boolean indexing

you can use boolean indexing with loc to!

but insted of an array you pass a series with the index as Series index and True/False values associated with each index value.
<br>
for example:

````python
df.Age>23
````

will result to:

````python
df.loc[df.Age>23]
````

you can chain many conditions with & and | opperators
<br>


````python
(df.Age>23)&(df.Age<25)
````

````python
df.loc[(df.Age>23)&(df.Age<25)]
````

you can still use select columns:

````python
df.loc[(df.Age>23)&(df.Age<25),["Age","Nationality"]]
````

### using MultiIndex

indexind by only one index could be very restricting <br> for example with Club index we can't choose a single row, but all the players in a club.

what if we want more intersections?
<br>
we can use MultiIndex!

with MultiIndex we can set an Heirarchecal index thet resembels the heirarchy of the data.
lets set index by columns

````python
df.set_index(["Nationality","Age","Name","ID"],drop=False)
````

````python
df
````

````python
df.sort_index(inplace=True)
````

````python
df.index.lexsort_depth
````

we can now use loc to query with MultiIndex
<br>
quering with the first level is a breeze

````python
df.loc["Spain"].head()
````

notice that you could query with the first and second levels like this:

````python
df.loc["Spain",23]
````

but this is not the recommanded way, because when you want to select a cerain column like this:
```python
df.loc["Spain",23,"Age"]
``` 
pandas won't be able to tell if by selecting Age you mean the 3rd index or the column Age <br>
the prefared way would be to use the loc usual loc sintax:
```python
df.loc[index,column]

```
but in the index you put a tuple with your index and after the comma the columns index:
```python
df.loc[(first_level,seconed_level,...,last_level),column]

```
for example:

````python
df.loc[("Spain",23),"Age"]
````

or:

````python
df.loc[("Spain",23),:].head()
````

what if you want to select all the players which are 23 years old from all Nationalitis?
you might try:
```python
df.loc[(:,23),:].head()
```
but this would result in an error (out of topic technical reasons)
the correct way to do this is using `slice()`

````python
df.loc[(slice(None),23),:]
````

### query

with query you can select data using query sintax
<br>
examples:

````python
df.query("Age==23").head()
````

````python
df.query("16<Age<19 and Nationality in ['Netherlands','Spain']").head()
````

you can also use variables availible at the current scope

````python
column_options = ['Netherlands','Spain']
````

````python
df.query("16<Age<19 and Nationality in @column_options").head()
````

you can even use the column name as a variable but the syntax is a bit funny

````python
column_name = "Nationality"
````

````python
df.query("16<Age<19 and @df[@column_name] in @column_options").head()
````
